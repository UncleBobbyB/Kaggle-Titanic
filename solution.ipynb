```python
print("Before", train_df.shape, test_df.shape, 
      combine[0].shape, combine[1].shape)

train_df = train_df.drop(['Ticket', 'Cabin'], axis=1)
test_df = test_df.drop(['Ticket', 'Cabin'], axis=1)
combine = [train_df, test_df]

print("After", train_df.shape, test_df.shape, combine[0].shape, 
        combine[1].shape)
```

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', 
                                                expand=False)
pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess', 
        'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 
            'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

train_df[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
title_mapping = {'Mr': 1, 'Miss': 2, 'Mrs': 3, 'Master': 4, 'Rare': 5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
```

```python
train_df = train_df.drop(['Name', 'PassengerId'], axis=1)
test_df = test_df.drop(['Name'], axis=1)
combine = [train_df, test_df]
print(train_df.shape, test_df.shape)
```

```python
train_df.head()
```

```python
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map({'female': 1, 
            'male': 0}).astype(int)
train_df.head()
```

```python
grid = sns.FacetGrid(train_df, row='Pclass', 
            col='Sex', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```python
guess_ages = np.zeros((2, 3))
guess_ages
```

```python
for dataset in combine:
    for i in range(0, 2):
        for j in range(0, 3):
            guess_df = dataset[(dataset['Sex'] == i) & \
                (dataset['Pclass'] == j + 1)]['Age'].dropna()
            age_guess = guess_df.median()
            guess_ages[i, j] = int(age_guess/0.5 + 0.5) * 0.5
    for i in range(0, 2):
        for j in range(0, 3):
            dataset.loc[(dataset.Age.isnull()) & (dataset.Sex == i) &\
                (dataset.Pclass == j + 1),\
                    'Age'] = guess_ages[i, j]
    dataset['Age'] = dataset['Age'].astype(int)

train_df.head()    
```

```python
train_df['AgeBand'] = pd.cut(train_df['Age'], 5)
train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], 
    as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
for dataset in combine:
    dataset.loc[dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[dataset['Age'] > 64, 'Age'] = 4

train_df.head()
```

```python
train_df = train_df.drop(['AgeBand'], axis=1)
combine = [train_df, test_df]
train_df.head()
```

```python
for dataset in combine:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
    
train_df[['FamilySize', 'Survived']].groupby(['FamilySize'], 
    as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 1, 'IsAlone'] = 1

train_df[['IsAlone', 'Survived']].groupby(['IsAlone'], as_index=False).mean()
```

```python
train_df = train_df.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
test_df = test_df.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
combine = [train_df, test_df]
```

```python
test_df['Fare'].fillna(test_df['Fare'].dropna().median(), inplace=True)
test_df.head()
```

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 4)
intervals = train_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False\
    ).mean().sort_values(by='FareBand', ascending=True)
intervals
```

```python
for dataset in combine:
    for i in range(4):
        dataset.loc[(dataset['Fare'] > intervals[i].left) &\
            (dataset['Fare'] <= intervals[i].right), 'Fare'] = i
    dataset['Fare'] = dataset['Fare'].astype(int)        
train_df = train_df.drop(['FareBand'], axis=1)
combine = [train_df, test_df]
```

```python
X_train = train_df.drop('Survived', axis=1)
Y_train = train_df['Survived']
X_test = test_df.drop('PassengerId', axis=1).copy()
print(X_train.shape, Y_train.shape, X_test.shape)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
Y_train.tolist()
```

```python
results = {}
```

```python
for name, cls in [['LogisticRegression', LogisticRegression], 
    ['SVC', SVC], ['LinearSVC', LinearSVC], 
    ['RandomForest', RandomForestClassifier], 
    ['KNeighbors', KNeighborsClassifier], ['GaussianNB', 
                   GaussianNB], ['Perceprton', Perceptron], 
                   ['SGD', SGDClassifier], 
                   ['DecisionTree', DecisionTreeClassifier]]:
    model = cls()
    model.fit(X_train, Y_train)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    train_accuracy = round(accuracy_score(Y_train.tolist(), train_pred) * 100, 
                          2)
    results[name] = train_accuracy
    print(name, ": ", 'train predictions accuracy: ', train_accuracy)
```

```python
models = {'Model': [], 'Score': []}
predictions = {}
for name, clf in [['Support Vector Machines', SVC], ['KNN', KNeighborsClassifier], 
    ['Logistic Regression', LogisticRegression], 
    ['Random Forest', RandomForestClassifier], ['Naive Bayes', GaussianNB], 
    ['Perceptron', Perceptron], 
    ['Stochastic Gradient Decent', SGDClassifier], [
        'Linear SVC', LinearSVC], 
        ['Decision Tree', DecisionTreeClassifier]]:
    model = clf()
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    predictions[name] = Y_pred
    accuracy = round(model.score(X_train, Y_train) * 100, 2)
    models['Model'].append(name)
    models['Score'].append(accuracy)
models
```

```python
models = pd.DataFrame(models)
models.sort_values(by='Score', ascending=False)
```

```python
param_grid = {'min_impurity_decrease': np.linspace(0, 0.2, 50)}

clf = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
clf.fit(X_train, Y_train)

print('best params: ', clf.best_params_, ' ', 'best score: ', 
     clf.best_score_)
```

```python
dtc = DecisionTreeClassifier(min_impurity_decrease=\
    clf.best_params_['min_impurity_decrease'])
# dtc = DecisionTreeClassifier()
dtc.fit(X_train, Y_train)
train_accuracy = round(dtc.score(X_train, Y_train) * 100, 2)
Y_pred2 = dtc.predict(X_test)

print('train accuracy: ', train_accuracy)
```

```python
entropy_thresholds = np.linspace(0, 1, 100)
gini_thresholds = np.linspace(0, 0.2, 100)
param_grid = [{'criterion': ['entropy'], 'min_impurity_decrease': 
    entropy_thresholds}, {'criterion': ['gini'], 'min_impurity_decrease': 
    gini_thresholds}, {'max_depth': np.arange(2, 10)}, 
    {'min_samples_split': np.arange(2, 30, 2)}]
clf = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
clf.fit(X_train, Y_train)
print('best params: ', clf.best_params_, ' ', 'best score: ', 
     clf.best_score_)
```
